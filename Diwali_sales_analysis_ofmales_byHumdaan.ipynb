```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as mpl
%matplotlib inline
import seaborn as sns
```

```python
df=pd.read_csv("Diwali Sales Data.csv",encoding='Unicode_escape')
df.head()
```

```python
df.drop(['Status','unnamed1'],axis=1)
```

```python
df.dropna()
df
```

```python
#Sorting by decreasing order of Sales
df.sort_values(by=['Amount'],ascending=False)
df 
```

```python
#grouping by gender

graph=sns.countplot(data=df,x='Gender',palette="Set3")
for bars in graph.containers:
    graph.bar_label(bars)
#
```

```python
# 7842 Orders or sales were places by females and 3409 orders were done by males

age_male_df=df.query('Gender == "M"')
age_male_df.head()
age_males_graph = sns.countplot(data=age_male_df,x='Age Group',palette='Set3')
for bar in age_males_graph.containers:
    age_males_graph.bar_label(bar)
```

```python
# Males of age group 26-35 placed most orders 

# Now  finding ocupation of males.
# age_male_df.groupby(by=['Occupation']).head()
graph=sns.countplot(data=age_male_df,x="Occupation")
graph.set_xticklabels(graph.get_xticklabels(), rotation=90)
```

```python
#It is found taht most of the men are from I.T sector

# Now checking these men are married or unmarried
graph=sns.countplot(data=age_male_df,x='Marital_Status',palette="Set3")
for bars in graph.containers:
    graph.bar_label(bars)
```

```python
# It is found that most Order placing men are unmarried.

#Now Finding which orders are they placing
sales_of_Unmarried_men = age_male_df.groupby(['Product_Category'], as_index=False)['Amount'].sum()
sales_of_Unmarried_men.head()
graph=sns.barplot(data=sales_of_Unmarried_men,x='Product_Category',y='Amount',palette="Set3")
graph.set_xticklabels(graph.get_xticklabels(), rotation=90)
```

```python
# Men placed most of the orders on Food

# finding which state tehy belong to
# age_male_df.info()
age_male_state_sales=age_male_df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
age_male_state_sales.head()
graph=sns.barplot(data=age_male_state_sales,x='State',y='Amount')
graph.set_xticklabels(graph.get_xticklabels(), rotation=90)
```

```python
# It is Found that most of the orders placed by unamrried men were from Uttarpradesh.
```

```python
### Final Conclusion
# 1.Females placed double the amount of orders as Males.
# 2.Males of age group 26-35 placed most orders. 
# 3.Most of these Males are from I.T sector.
# 4.Martial status of these men are Unmarried.
# 4.These Men spent most amount of money on Food
# 5.These men resided in Uttarpradesh.
```
